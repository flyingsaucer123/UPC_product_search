```python
!pip install crewai langchain-google-genai crewai_tools

import os
from dotenv import load_dotenv
from crewai import Task
load_dotenv()

from crewai import Agent
from langchain.tools import tool
from langchain_google_genai import ChatGoogleGenerativeAI

# Setting up Google Gemini Flash
llm = ChatGoogleGenerativeAI(
    model="gemini-1.5-flash",
    verbose=True,
    temperature=0.5,
    google_api_key="your_gemini_key"  # Use your API key here
)

# Define the UPC Search Agent
upc_search_agent = Agent(
    role="UPC Search Specialist",
    goal="Search for the UPC code based on the product description, manufacturer, and part number.",
    backstory=("You are skilled at retrieving UPC codes based on product data from various online sources."),
    tools=[],
    llm=llm,
    allow_delegation=True,
    verbose=True,
    memory=True
)

# Define the UPC Verification Agent
upc_verification_agent = Agent(
    role="UPC Verification Expert",
    goal="Verify the UPC code retrieved from multiple sources for accuracy.",
    backstory=("You specialize in validating UPC codes using multiple databases and tools."),
    tools=[],
    llm=llm,
    allow_delegation=False,
    verbose=True,
    memory=True
)

# Adding Serper tool for searching
from crewai_tools import SerperDevTool

os.environ["SERPER_API_KEY"] = "your_serper_key"  # Serper API key

tool = SerperDevTool()

# UPC Search Task
upc_search_task = Task(
    description=(
        "Use the product details (manufacturer, product description, part number) "
        "to search for the corresponding UPC code."
    ),
    expected_output='UPC code retrieved based on product details.',
    tools=[tool],  # Using Serper for the search task
    agent=upc_search_agent
)

# UPC Verification Task
upc_verification_task = Task(
    description=(
        "Verify the UPC code by cross-checking it across multiple sources to ensure accuracy."
    ),
    expected_output='Verified UPC code from multiple sources.',
    tools=[tool],  # Using Serper for verification as well
    agent=upc_verification_agent
)

# Create the crew for the UPC retrieval process
from crewai import Crew, Process

crew = Crew(
    agents=[upc_search_agent, upc_verification_agent],
    tasks=[upc_search_task, upc_verification_task],
    process=Process.sequential  # Running the tasks in sequence
)

# Example product data to search for UPC
result = crew.kickoff(inputs={
    "manufacturer_name": "ABCO PRODUCTS",
    "product_description": "32 GAL ROUND WASTE RECEPTACLE",
    "manufacturer_part_number": "BS32G"
})

print(result)
```
